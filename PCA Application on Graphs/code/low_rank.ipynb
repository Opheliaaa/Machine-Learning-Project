```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread
import scipy
%matplotlib inline 
```

# Part a)

```python
def low_rank_approximation(X,rank):
    # YOUR CODE GOES HERE
    u, s, vt = np.linalg.svd(X,full_matrices=False)
    s[rank:]=0
    return u @ np.diag(s) @ vt
```

```python
face = imread("./data/face.jpg")
plt.imshow(face)
```

```python
rank5=low_rank_approximation(face,5)
plt.imshow(rank5)
```

```python
rank20=low_rank_approximation(face,20)
plt.imshow(rank20)
```

```python
rank100=low_rank_approximation(face,100)
plt.imshow(rank100)
```

# Part c)

```python
def mse(img1, img2):
    # YOUR CODE GOES HERE 
    residual=np.square(np.subtract(img1,img2))
    return np.sum(residual)
```

```python
#face
mse_list=[]
for i in range(1,101):
    mse_list.append(mse(face,low_rank_approximation(face,i)))
```

```python
#sky
mse_list2=[]
for i in range(1,101):
    mse_list2.append(mse(sky,low_rank_approximation(sky,i)))
```

```python
plt.plot(range(1,101),mse_list,label='face_mse')
plt.plot(range(1,101),mse_list2,label='sky_mse')
plt.xlabel('rank')
plt.ylabel('MSE')
plt.legend()
plt.title('image of face')
plt.show()
```

# Part d)

For face image, original and the approximated images become hard to differentiate when rank roughly larger than 50.
For sky image, original and the approximated images become hard to differentiate when rank roughly larger than 5.
The difference can be due to the complexity of the image. Face image is much more complex than the sky matrix. Even pixels in face image that are around each other are very different, so that's why low rank approximation isn't as good as the sky matrix given the same rank. The original rank of face image may be larger than the original rank of sky.
